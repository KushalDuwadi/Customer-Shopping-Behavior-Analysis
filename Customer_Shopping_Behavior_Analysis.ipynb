```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.columns = df.columns.str.replace(' ' ,'_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-ages', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(20)
```

```python
# create column purchae_frequency_days
# in our data set there is data like 7,14,365,90,... as purchae_frequency_days so transforming then into numerical value for analysis.

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Quarterly' : 90,
    'Monthly':30,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['purchase_frequency_days'].value_counts()
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
